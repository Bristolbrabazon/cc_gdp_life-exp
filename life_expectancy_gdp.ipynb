```python
import pandas as pd
import numpy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('all_data.csv')
print(data.head())
```

```python
# Has life expectancy increased over time in the six nations?
sns.lineplot(data=data, y='Life expectancy at birth (years)', x='Year', hue='Country')
plt.show()
plt.clf()
```

```python
# Has GDP increased over time in the six nations?
sns.lineplot(data=data, y='GDP', x='Year', hue='Country')
plt.show()
plt.clf()
```

```python
# Is there a correlation between GDP and life expectancy of a country?
sns.scatterplot(data=data, y='Life expectancy at birth (years)', x='GDP', hue='Country')
plt.xscale('log')
plt.show()
plt.clf()
```

```python
# What is the average life expectancy in these nations?
sns.boxplot(data=data, y='Life expectancy at birth (years)', x='Country')
plt.xticks(rotation=30)
plt.show()
plt.clf()
```
